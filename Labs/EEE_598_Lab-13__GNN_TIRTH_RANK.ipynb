```python
"""
Module 2.2: Function to Visualize the Graph Structure
- Passing the adjacency matrix, original node mask, and neighborhood node mask
- Original node colored brown, neighborhood nodes colored orange, others white
"""

def draw_graph_structure(adjacency_matrix, original_node, neighborhood_nodes=None):

  G = nx.Graph()
  n_node = adjacency_matrix.shape[0]
  for i in range(n_node):
    for j in range(i):
      if adjacency_matrix[i,j]:
          G.add_edge(i,j)

  color_map = []

  for node in G:
    if original_node[node]:
      color_map.append('brown')
    else:
      if neighborhood_nodes[node]:
        color_map.append('orange')
      else:
        color_map.append('white')

  nx.draw(G, nx.spring_layout(G, seed = 7), with_labels=True,node_color=color_map)
  plt.show()
```

```python
# --- Simple GNN Implementation Example ---
# This shows how the same idea (neighbors -> aggregation -> MLP) works in a real model

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv

# --- Step 1: Convert your adjacency matrix to edge_index format (PyTorch Geometric expects COO format) ---
edge_index = torch.tensor(np.array(np.nonzero(A)), dtype=torch.long)

# --- Step 2: Define node features ---
# For teaching, we'll just use random features for each node
x = torch.randn((A.shape[0], 5))  # 5-dimensional feature vector for each node

# --- Step 3: Create a small GNN model ---
class SimpleGCN(nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(SimpleGCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        # 1st layer: each node aggregates info from direct neighbors
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        # 2nd layer: info flows two hops away
        x = self.conv2(x, edge_index)
        return x

# --- Step 4: Create graph data object ---
data = Data(x=x, edge_index=edge_index)

# --- Step 5: Initialize model and run forward pass ---
model = SimpleGCN(in_channels=5, hidden_channels=8, out_channels=2)
out = model(data.x, data.edge_index)

print("Node embedding output shape:", out.shape)
print("Example node embeddings:\n", out)
```
